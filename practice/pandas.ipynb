## panda series

### A Pandas Series is like a column in a table.

*It is a one-dimensional array holding data of any type.*

```python
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

- **With the index argument, you can name your own labels.**

```python
a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
```

```python
print(myvar["y"])
```

```python
lst = [1,2,3,4,5,6]
print(lst)
series = pd.Series(lst)
print(series)
print(type(series))
```

```python
a = pd.Series(['p','q','r','s','t'], index = [10,11,12,13,14])
a
```

```python
b = pd.Series(['p','q','r','s','t'], index = [10,11,12,13,14], name = 'alphabets')
b
```

```python
scalar_series = pd.Series(0.5, index = [1,2,3])
scalar_series
```

- **You can also use a key/value object, like a dictionary, when creating a Series.**

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

- **The keys of the dictionary become the labels.**

```python
print(myvar['day1'])
```

- **To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series.**

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
```

```python
dict_series = pd.Series({'p':[1,5,6], 'q':[2,6,7], 'r':[3,9,0], 's':[4,4,5], 't':[5,1,2]})
dict_series
```

## Data Frames

```python
df = pd.DataFrame()
print(df)
```

```python
lst = [1,2,3,4,5]
df = pd.DataFrame(lst)
df
```

```python
lst = [[1,2,3,4,5], [11,12,13,14,15]]
df = pd.DataFrame(lst)
df
```

## Merge() Function

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Physics':[34,67,34,89,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Chemistry':[78,33,39,81,90]})
df2
```

```python
pd.merge(df1, df2, on = 'Roll No.') 
```

```python
pd.merge(df2, df1, on = 'Roll No.') 
```

```python
pd.merge(df1, df2)   # on - by default intersecting column is taken
```

```python
df3 = pd.DataFrame({'Roll No.':[1,2,3,6,7], 'Physics':[34,67,34,89,12]})
df4 = pd.DataFrame({'Roll No.':[1,2,3,8,9], 'Chemistry':[34,67,34,89,12]})
```

```python
pd.merge(df3, df4)
```

```python
pd.merge(df3, df4, how = 'left')
```

```python
pd.merge(df3, df4, how = 'right')
```

```python
pd.merge(df3, df4, how = 'outer')
```

## concat() Function

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Maths':[45,78,45,90,66], 'Physics':[33,67,12,90,44]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10], 'Maths':[78,73,45,90,69], 'Physics':[23,67,88,0,98]})
print(df1)
print(df2)
```

```python
df3 = pd.concat([df1, df2], ignore_index=True)
print(df3)
```
